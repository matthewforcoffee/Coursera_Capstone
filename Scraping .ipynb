```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import requests
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

### So first step was to read the data and convert it to numpy array. Then i found the Boroughs which are unassigned and deleted them.

### I had to separate the columns because it was in a list with multiple different columns lengths and couldn't join it back.

##### I finally rejoined everything and was able to get the table

```python
dfs = pd.read_html(url)
df = np.array(dfs)
df = df[0]
df2 = df.to_numpy()
col_1 = []
col_2 = []
col_3 = []
for i in range(1,len(df2)):
    if df2[i,1] != 'Not assigned':
        selectrow = df2[i,:]
        #print(selectrow)
        col1 = [selectrow[0]]
        #print(col1)
        col2 = [selectrow[1]]
        #print(col2)
        col3 = selectrow[2:]
        #print(col3)     
        col_1 = col_1 + col1
        col_2 = col_2 + col2
        col_3 = np.concatenate((col_3,col3))

join1= np.vstack((col_1,col_2)).transpose()
joined2 = np.column_stack((join1,col_3))
final_table = pd.DataFrame(joined2, columns = ["Postal Code", "Borough", "Neighborhood"])
final_table
```

```python
!wget -q -O 'Geospatial_Coordinates.csv' http://cocl.us/Geospatial_data
geospatial_coordinates = pd.read_csv('Geospatial_Coordinates.csv')

y = pd.DataFrame(columns = ['Postal Code','Borough','Neighborhood','Latitude','Longitude'])
for i in range(1,len(final_table)):
    for j in range(1,len(geospatial_coordinates)):
        if final_table.iloc[i,0] == geospatial_coordinates.iloc[j,0]:
            final_table2 = pd.concat([final_table.iloc[i,:], geospatial_coordinates.iloc[j,1:]])           
            y= y.append(final_table2, ignore_index = True)
print(y)
```

# Part 3: Clustering the Neighborhoods in Toronto

```python
!conda install -c conda-forge folium=0.5.0 --yes
```

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
import folium # map rendering library

latitude = 43.5
longitude = -79.4
toronto_grouped = y.groupby('Neighborhood').mean().reset_index()

# set number of clusters
kclusters = 4

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(y['Latitude'], y['Longitude'], y['Neighborhood'], kmeans.labels_):
    label = folium.Popup(str(poi) + ' Cluster' + str(kmeans), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
