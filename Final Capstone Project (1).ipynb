    1. Is it in a mall
    2. Is it near a centroid of a cluster
    3. is it far from the next haidilao or beauty in the pot?

The evaluation of the prediction will then be done using the jaccard similarity score.

```python
criteria1 = dataframe2.append(dataframe4)
criteria1['Criteria 1'] = True
criteria1.reset_index(inplace=True, drop=True)
criteria1
```

From the DataFrame, the address of these franchises are all in malls, therefore the column 'In a Mall' is all True

```python
df_group3 = df_grouped3
df_group3['lat'] = pd.to_numeric(df_group3['lat'])
df_group3['lng'] = pd.to_numeric(df_group3['lng'])
print(df_group3.dtypes)

criteria2 = df_group3.groupby('Clus_Db', as_index = False).mean()
criteria2
```

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lat1, lon1, lat2, lon2):

      R = 3959.87433 # this is in miles.  For Earth radius in kilometers use 6372.8 km

      dLat = radians(lat2 - lat1)
      dLon = radians(lon2 - lon1)
      lat1 = radians(lat1)
      lat2 = radians(lat2)

      a = sin(dLat/2)**2 + cos(lat1)*cos(lat2)*sin(dLon/2)**2
      c = 2*asin(sqrt(a))

      return R * c

center_point = pd.DataFrame([criteria1.loc[:,'lat'],  criteria1.loc[:,'lng']])
center_point = center_point.T
test_point = pd.DataFrame([criteria2.loc[:,'lat'],  criteria2.loc[:,'lng'], criteria2.loc[:,'Clus_Db']])
test_point = test_point.T

trainset_eval = criteria1

for traindata2 in range(0,len(criteria1)):
    lat1 = center_point.iloc[traindata2,0]
    lon1 = center_point.iloc[traindata2,1]
    for clusterdata in range(0,len(criteria2)):
        lat2 = test_point.iloc[clusterdata,0]
        lon2 = test_point.iloc[clusterdata,1]
        radius = 2.50 # in kilometer

        a = haversine(lon1, lat1, lon2, lat2)

        #print('Distance (km) : ', a)
        if a <= radius:
            #print('Inside cluster ' + str(test_point.iloc[clusterdata,2]))
            trainset_eval.loc[traindata2,'Near a cluster'] = str(test_point.iloc[clusterdata,2])
        else:
            #print('Outside the cluster')
            pass
```

```python
trainset_eval['Near a cluster'] = trainset_eval['Near a cluster'].astype(float)
for yesno in range(0,len(trainset_eval)):
    if trainset_eval.loc[yesno,'Near a cluster'] > 0:
        trainset_eval.loc[yesno,'Criteria 2'] = True
    else:
        trainset_eval.loc[yesno,'Criteria 2'] = False
trainset_eval2 = trainset_eval.drop(['ID', 'Formatted Address'],axis=1)

trainset_eval2
```

```python
xx = trainset_eval2.loc[:,['lat','lng','Criteria 1','Criteria 2']]
trainset_eval2['Success'] = 1
yy = trainset_eval2.loc[:,['Success']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xx, yy, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Train set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
decTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
decTree.fit(X_train,y_train)
predTree = decTree.predict(X_test) #Predict the decisions
from sklearn import metrics
import matplotlib.pyplot as plt

print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

So using the decision tree prediction, the data seems to be accurate in giving a Yes or No decision on whether it should be set up by fulfilling the criteria.
